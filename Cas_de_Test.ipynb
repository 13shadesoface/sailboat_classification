```python
import matplotlib.pyplot as plt
import numpy as np
import sys
import numpy as np
import perceptron
```

## Classification

### Linear Simple :
        Linear Model : OK
        MLP (2, 1)   : OK

#### Data

```python
X = np.array([
      [1., 1.],
      [2., 3.],
      [3., 2.]
])
Y = np.array([
              [1],
              [-1],
              [-1],
])
```

#### Création du modèle

```python
model = perceptron.MLP([2, 1])
```

#### Before training

```python
for sample_input in X:
    print(model.predict(sample_input))
```

#### After training

```python
model.train(X, Y)
for sample_input in X:
    print("sample input", sample_input)
    print("prediction", model.predict(sample_input))
```

#### Représentation modèle

```python
predict_x1 = np.arange(1, 3, 0.01)
predict_x2 = np.arange(1, 3, 0.01)
px_1 = []
px_2 = []
predicted_color = []
for x1 in predict_x1:
  for x2 in predict_x2:
    px_1.append(x1)
    px_2.append(x2)
    predicted_color.append('lightblue' if model.predict(np.array([x1, x2]))[0] >= 0 else 'lightpink')
```

```python
plt.scatter(px_1, px_2, c=predicted_color)
plt.scatter(X[0, 0], X[0, 1], color='blue')
plt.scatter(X[1:3,0], X[1:3,1], color='red')
plt.show()
plt.clf()
```

### Multi Linear 3 classes :
        Linear Model x3 : OK
        MLP (2, 3)      : OK

```python
X = np.random.random((500, 2)) * 2.0 - 1.0
Y = np.array([[1, 0, 0] if -p[0] - p[1] - 0.5 > 0 and p[1] < 0 and p[0] - p[1] - 0.5 < 0 else 
              [0, 1, 0] if -p[0] - p[1] - 0.5 < 0 and p[1] > 0 and p[0] - p[1] - 0.5 < 0 else 
              [0, 0, 1] if -p[0] - p[1] - 0.5 < 0 and p[1] < 0 and p[0] - p[1] - 0.5 > 0 else 
              [0, 0, 0]for p in X])

X = X[[not np.all(arr == [0, 0, 0]) for arr in Y]]
Y = Y[[not np.all(arr == [0, 0, 0]) for arr in Y]]

print(Y)
```

```python
model = perceptron.MLP([2, 3])
```

```python
model.train(X, Y, learning_rate=0.001, nb_iter=100000)
```

```python
predict_x1 = np.arange(-1, 1, 0.01)
predict_x2 = np.arange(-1, 1, 0.01)
px_1 = []
px_2 = []
predicted_color = []
for x1 in predict_x1:
  for x2 in predict_x2:
    px_1.append(x1)
    px_2.append(x2)
    prediction = model.predict(np.array([x1, x2]))
    if max(prediction[0], prediction[1], prediction[2]) == prediction[0]:
        predicted_color.append('lightblue')
    elif max(prediction[0], prediction[1], prediction[2]) == prediction[1]:
        predicted_color.append('lightpink')
    elif max(prediction[0], prediction[1], prediction[2]) == prediction[2]:
        predicted_color.append('lightgreen')
```

```python
plt.scatter(px_1, px_2, c=predicted_color)

plt.scatter(np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]][0] == 1, enumerate(X)))))[:,0], np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]][0] == 1, enumerate(X)))))[:,1], color='blue')
plt.scatter(np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]][1] == 1, enumerate(X)))))[:,0], np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]][1] == 1, enumerate(X)))))[:,1], color='red')
plt.scatter(np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]][2] == 1, enumerate(X)))))[:,0], np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]][2] == 1, enumerate(X)))))[:,1], color='green')
plt.show()
plt.clf()
```
